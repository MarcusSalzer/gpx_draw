## saving

```python
# Save pandas json
act_dfs = {}
for k in act_dicts.keys():
    p = act_dicts[k]["points"]
    df = pd.DataFrame.from_dict(p).fillna(pd.NA)
    df["time"] = pd.to_datetime(df["time"], cache=True)
    df.reset_index(drop=True, inplace=True)
    act_dfs[k] = df
```

```python
for k in act_dicts.keys():
    df = act_dfs[k]
    df.to_json("../data/points_pandas/" + k + ".json")
```

```python
act_dfs_loaded = {}
for k in act_dicts.keys():
    a = pd.read_json("../data/points_pandas/" + k + ".json", convert_dates=["time"])
    act_dfs_loaded[k] = a
```

```python
for k in act_dfs.keys():
    print(all(act_dfs[k].index == act_dfs_loaded[k].index), end=" ")
print()
for k in act_dfs.keys():
    print(all(act_dfs[k].columns == act_dfs_loaded[k].columns), end=" ")
print()
for k in act_dfs.keys():
    print(all(act_dfs[k].dtypes == act_dfs_loaded[k].dtypes), end=" ")

print()
for c in act_dfs[k].columns:
    eq = True
    for k in act_dfs.keys():
        if not (act_dfs[k][c].equals(act_dfs_loaded[k][c])):
            eq = False
    print(f"{c:10s}", ":", eq)
```

## activities

```python
## Convert to act object
acts = [activity.Act.from_dict(d) for d in act_dicts.values()]
```

```python
display(acts[0].points)
```

```python
print("total points:", sum((len(a.points) for a in acts)))
```

```python
acts[0].points.head()
```

```python
# Create dict listing all points, activities separated by None
points_all = {k: [] for k in ["long", "lat", "time"]}
for act in acts:
    for k in points_all.keys():
        points_all[k].extend(act.points[k])
        points_all[k].append(None)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=points_all["long"], y=points_all["lat"]))
fig.update_layout(template="plotly_dark")
fig.update_yaxes(scaleanchor="x", scaleratio=1)
fig.show()
```
