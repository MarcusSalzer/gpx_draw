```python
from gpxpy import parse
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use("dark_background")
```

```python
file_path = "data/activity_14718672200.gpx"

with open(file_path, "r") as f:
    data = parse(f)

print("# tracks:", len(data.tracks))
print("# segments:", len(data.tracks[0].segments))
print("# points:", len(data.tracks[0].segments[0].points))
```

```python
s = data.tracks[0].segments[0]
points = s.points
lat = [p.latitude for p in points]
long = [p.longitude for p in points]
elev = [p.elevation for p in points]
time = [p.time for p in points]

coords = np.array([long, lat, elev]).T
print(coords.shape)

plt.figure()
plt.plot(coords[:, 0], coords[:, 1])
plt.show()
```
